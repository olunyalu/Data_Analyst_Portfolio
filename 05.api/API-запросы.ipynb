```python
# Подключение библиотек
import requests
import pandas as pd
from pprint import pprint
```


Существует четыре основных способа (или метода) запроса информации с
веб-сервера через `HTTP`:
- GET();
- POST();
- PUT();
- DELETE().


Для тестирования запросов к API JSONPlaceholder вы можете использовать любой HTTP-клиент, например Postman.

## GET-запросы

Для работы с API JSONPlaceholder вам потребуется отправлять запросы HTTP на указанный URL с определенными методами и параметрами.

    Данный запрос возвращает все существующие посты:

```python
response_1 = requests.get('https://jsonplaceholder.typicode.com/posts')
response_1.status_code
```

Переводим в формат json - в python это словарь.

```python
posts = response_1.json()
pprint(posts)
```

Опрелеляем списки для хранения данных:

```python
body = list()
id = list()
title = list()
userId= list()
dict_to_df = {}
```

Перебираем каждую запись и записываем полученные данные в словарь, чтобы сформировать датафрейм.

```python
for k in posts:
    body.append(k.get('body'))
    id.append(k.get('id'))
    title.append(k.get('title'))
    userId.append(k.get('userId'))
```

```python
dict_to_df = {'body': body,
              'id': id,
              'title': title,
              'userId': userId}
```

```python
pd.DataFrame(dict_to_df)
```

    Данный запрос возвращает конкретный пост по его ID.

```python
response_2 = requests.get('https://jsonplaceholder.typicode.com/posts?id=1')
print(response_2.status_code)
post = response_2.json()
pprint(post)
```

    Данный запрос возвращает 10 постов.

```python
response_3 = requests.get('https://jsonplaceholder.typicode.com/posts?_page=0&_limit=10')
print(response_3.status_code)
posts_10 = response_3.json()
pprint(posts_10)
```

## POST-запросы

    Данный запрос создает новый пост с указанными данными.

```python
url = 'https://jsonplaceholder.typicode.com/posts'
data = {
    'title': 'New post',
    'body': 'This is the body of the new post',
    'userId': 1
}
```

```python
response_4 = requests.post(url, json = data)
```

```python
if response_4.status_code == 201:
    print('Post created successfully')
else:
    print('Failed to create post')
```

```python
pprint(response_4.json())
```

    Данный запрос создает нового пользователя с указанными данными.

```python
url_ = 'https://jsonplaceholder.typicode.com/users'
data_ = {
    'name': 'Olya Gavrilova',
    'username': 'olunyalu',
    'email': 'theolenkapetrova@gmail.com'
}
```

```python
response_5 = requests.post(url_, json = data_)
```

```python
if response_5.status_code == 201:
    print('User created successfully')
else:
    print('Failed to create user')
```

```python
pprint(response_5.json())
```

### В отличие от метода GET, который используется для получения данных, POST служит для их отправки на сервер. При этом они передаются не с URL, а в теле HTTP-запроса.

## PUT-запросы

    Данный запрос обновляет существующий пост по его ID.

```python
link = 'https://jsonplaceholder.typicode.com/posts/1'
info =  {
    'title': 'Update post',
    'body': 'This is the body of the update post',
    'userId': 1
}
response_6 = requests.put(link, json = info)
print(response_6.status_code)
pprint(response_6.json())
```

## DELETE-запросы

        Данный запрос удаляет существующий пост по его ID.

```python
response_7 = requests.delete('https://jsonplaceholder.typicode.com/posts/1')
print(response_7.status_code)
pprint(response_7.json())
```
